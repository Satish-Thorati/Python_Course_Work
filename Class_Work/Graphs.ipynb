**Graphs**

**1**. Line Plot
| **Attribute**       | **Description**                                  | **Example Code**                                                                                                                     |
| ------------------- | ------------------------------------------------ | ------------------------------------------------------------------------------------------------------------------------------------ |
| **color**           | Sets the line color                              | `plt.plot(x, y, color="red")`                                                                                                        |
| **marker**          | Adds markers for each data point                 | `plt.plot(x, y, marker="o")` → circle <br> `plt.plot(x, y, marker="s")` → square <br> `plt.plot(x, y, marker="^")` → triangle        |
| **linestyle (ls)**  | Style of the line connecting data points         | `plt.plot(x, y, linestyle="-")` → solid <br> `plt.plot(x, y, linestyle="--")` → dashed <br> `plt.plot(x, y, linestyle=":")` → dotted |
| **linewidth (lw)**  | Thickness of the line                            | `plt.plot(x, y, linewidth=3)`                                                                                                        |
| **label**           | Adds a legend entry (used with `plt.legend()`)   | `plt.plot(x, y, label="Stock A")` <br> `plt.legend()`                                                                                |
| **alpha**           | Controls transparency (0 = invisible, 1 = solid) | `plt.plot(x, y, alpha=0.5)`                                                                                                          |
| **grid**            | Adds grid lines to the plot                      | `plt.grid(True)` <br> `plt.grid(linestyle=":")`                                                                                      |
| **title**           | Title of the plot                                | `plt.title("My Graph")`                                                                                                              |
| **xlabel / ylabel** | Labels for x and y axes                          | `plt.xlabel("X Axis")` <br> `plt.ylabel("Y Axis")`                                                                                   |


```python
%pip install matplotlib
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5, 6, 7, 8]
visitors = [50, 60, 55, 70, 90, 120, 110, 130]

plt.plot(hours, visitors,
         color="green",
         marker="o",
         linestyle="-",
         linewidth=1,
         alpha=0.9,
         label="Visitors")

plt.title("Website Visitors Per Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Visitors")
plt.legend()

plt.grid(True,linestyle="-", alpha=0.1)
plt.show()
```

### 2. Bar Chart

| Attribute     | Description              | Example                                |
|--------------|--------------------------|----------------------------------------|
| **color**     | Bar color                | `plt.bar(x, y, color="red")`           |
| **width**     | Thickness of the bar     | `plt.bar(x, y, width=0.5)`             |
| **label**     | Used for legend          | `plt.bar(x, y, label="Sales")`         |
| **alpha**     | Transparency             | `plt.bar(x, y, alpha=0.7)`             |
| **edgecolor** | Border color of bars     | `plt.bar(x, y, edgecolor="black")`     |


```python
import matplotlib.pyplot as plt

products = ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch"]
sales = [250, 400, 150, 300, 180]

plt.bar(products, sales, color="blue",edgecolor='pink',alpha=0.5,width=0.1)

plt.title("Product Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Units Sold")

plt.show()
```

### 3. Histogram

| Attribute     | Description                         | Example                                         |
|--------------|-------------------------------------|-------------------------------------------------|
| **bins**      | Number of intervals                 | `bins=10`                                       |
| **color**     | Fill color                          | `color="green"`                                 |
| **edgecolor** | Border color of bins                | `edgecolor="black"`                             |
| **alpha**     | Transparency                        | `alpha=0.7`                                     |
| **histtype**  | Style of histogram                  | `"bar"`, `"step"`, `"stepfilled"`               |
| **density**   | Normalize values into probabilities | `density=True`                                  |


```python
import matplotlib.pyplot as plt

marks = [45, 55, 60, 72, 68, 80, 90, 55, 67, 70,
         85, 40, 60, 75, 78, 82, 95, 50, 62, 65,
         58, 73, 77, 69, 71, 88, 92, 55, 63, 59]

plt.hist(marks, bins=20, histtype='stepfilled',color="skyblue", edgecolor="black")

plt.title("Distribution of Student Marks")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")

plt.show()
```

### 4. Scatter Plot

| Attribute    | Description                             | Example                                              |
|-------------|-----------------------------------------|------------------------------------------------------|
| `x, y`       | Data for X and Y axes                   | `plt.scatter(ad_spend, sales)`                       |
| `s`          | Size of markers (single value or list)  | `s=200` or `s=[100,200,300]`                         |
| `c`          | Color of markers (single value or list) | `c="red"` or `c=sales`                               |
| `cmap`       | Colormap when `c` is numeric            | `cmap="plasma"`                                      |
| `alpha`      | Transparency of points (0 to 1)         | `alpha=0.7`                                          |
| `marker`     | Shape of marker                         | `"o"` (circle), `"s"` (square), `"^"` (triangle)     |
| `edgecolors` | Border color of markers                 | `edgecolors="black"`                                 |
| `linewidths` | Thickness of marker edges               | `linewidths=1.5`                                     |


```python
import matplotlib.pyplot as plt

ad_spend = [2, 4, 5, 7, 8, 10, 11, 12]
sales = [27, 25, 28, 35, 40, 48, 50, 52]

sizes = [200, 250, 90, 350, 40, 80, 500, 520]

colors = [1,2,3,4,5,6,7,8]

plt.scatter(ad_spend, sales,
            s=sizes,
            c=colors,
            cmap='plasma',
            alpha=0.8,
            edgecolors='black',
            linewidths=1.2
           )

plt.title("Advertising Spend vs Sales (with Color & Size)", fontsize=14, fontweight="bold")
plt.xlabel("Ad Spend ($1000s)", fontsize=12)
plt.ylabel("Sales ($1000s)", fontsize=12)

plt.colorbar(label="Sales Value")

plt.grid(True, linestyle="--", alpha=0.2)
plt.show()
```

**5. Pie Chart**

```python
import matplotlib.pyplot as plt

brands = ["Apple", "Samsung", "Xiaomi", "Oppo", "Others"]
market_share = [30, 25, 20, 10, 15]

plt.pie(market_share,
        labels=brands,
        autopct="%1.2f%%",
        startangle=90,
        colors=["gold", "skyblue", "lightcoral", "lightgreen", "violet"],
        explode=[0, 0, 0, 0.3, 0],
        shadow=True)

plt.title("Mobile Market Share 2025")
plt.show()
```

### 6. Box Plot

| Attribute      | Description                                      | Example                                                      |
|---------------|--------------------------------------------------|--------------------------------------------------------------|
| `x`            | Data to plot (list/array of datasets)            | `plt.boxplot([data1, data2])`                                |
| `labels`       | Labels for each dataset                          | `labels=["Class A", "Class B"]`                              |
| `vert`         | Orientation: `True`=vertical, `False`=horizontal | `vert=False`                                                 |
| `notch`        | Shows notch around median                        | `notch=True`                                                 |
| `patch_artist` | Fills the boxes with color                       | `patch_artist=True`                                          |
| `meanline`     | Shows mean line instead of median                | `meanline=True`                                              |
| `showmeans`    | Displays mean marker                             | `showmeans=True`                                             |
| `boxprops`     | Style for boxes (color, linewidth, etc.)         | `boxprops=dict(color="blue", facecolor="lightblue")`         |
| `medianprops`  | Style for median line                            | `medianprops=dict(color="red", linewidth=2)`                 |
| `whiskerprops` | Style for whiskers                               | `whiskerprops=dict(color="green")`                           |
| `capprops`     | Style for caps                                   | `capprops=dict(color="black")`                               |
| `flierprops`   | Style for outliers                               | `flierprops=dict(marker="o", color="orange", alpha=0.5)`     |


```python
import matplotlib.pyplot as plt

scores = [
    [70, 85, 90, 92, 88, 76, 95, 9],
    [65, 70, 68, 72, 75, 80, 85, 78],
    [90, 92, 94, 96, 88, 91, 89, 93],
    [62, 60, 61, 90, 65, 63, 67, 61],
    [70, 85, 8, 50, 92, 91, 100, 89]
]
plt.boxplot(scores,
            patch_artist=False,
            notch=True,
            vert=False,
            labels=["A", "B", "C", "D", "E"])

plt.title("Exam Scores Distribution Across Classes")
plt.xlabel("Classes")
plt.ylabel("Scores")
plt.grid(True, linestyle="--", alpha=0.7)

plt.show()
```

### 7. Area Chart

| Attribute   | Description                                  | Example                                |
|------------|----------------------------------------------|----------------------------------------|
| `x, y`      | Data for X and Y axes                        | `plt.fill_between(days, visitors)`     |
| `color`     | Fill color of the area                       | `color="skyblue"`                      |
| `alpha`     | Transparency of the filled area (0 to 1)     | `alpha=0.5`                            |
| `label`     | Label for legend                             | `label="Visitors"`                     |
| `linewidth` | Line thickness (when combined with `plot()`) | `linewidth=2`                          |
| `linestyle` | Style of line (when combined with `plot()`)  | `linestyle="--"`                       |

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
visitors = [120, 150, 170, 200, 180, 220, 250]

plt.fill_between(days, visitors,
                 color="skyblue", alpha=0.5,
                 label="Website Visitors")

plt.plot(days, visitors, color="blue", linewidth=2)
plt.title("Website Visitors Over a Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Visitors")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### 8. Heatmap (Seaborn)

| Attribute     | Description                      | Example                                  |
|--------------|----------------------------------|------------------------------------------|
| `data`        | 2D data (list, array, dataframe) | `np.array([...])`                        |
| `annot`       | Show values inside cells         | `annot=True`                             |
| `fmt`         | Format for annotations           | `fmt="d"` (int), `fmt=".1f"` (float)     |
| `cmap`        | Color map for intensity          | `cmap="coolwarm"`                        |
| `linewidths`  | Space/line between cells         | `linewidths=0.5`                         |
| `xticklabels` | Labels for X-axis                | `xticklabels=days`                       |
| `yticklabels` | Labels for Y-axis                | `yticklabels=weeks`                      |
| `cbar`        | Show/hide color bar              | `cbar=False`                             |


```python
%pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.array([
    [30, 32, 31, 29, 28, 27, 26],
    [33, 34, 32, 30, 29, 28, 27],
    [35, 36, 34, 33, 31, 30, 29],
    [37, 38, 35, 34, 32, 31, 30],
    [36, 37, 34, 32, 31, 30, 29]
])

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

weeks = ["Week 1", "Week 2", "Week 3", "Week 4", "Week 5"]

sns.heatmap(data,
            annot=True,
            fmt="d",
            cmap="YlOrRd",
            linewidths=0.5,
            xticklabels=days,
            yticklabels=weeks)

plt.title("Weekly Temperature Heatmap", fontsize=14, fontweight="bold")
plt.xlabel("Days of the Week")
plt.ylabel("Weeks")
plt.show()
```

**9. Subplots**

```python
#All Sub Plots
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
sales = [120, 135, 150, 160, 170, 200, 220]
profit = [30, 40, 45, 50, 55, 70, 80]

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(days, sales, marker="o", color="blue")
plt.title("Sales Over Days")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.bar(days, sales, color="orange")
plt.title("Sales Bar Chart")

plt.subplot(2, 2, 3)
plt.scatter(sales, profit, color="green", s=100, edgecolors="black")
plt.title("Sales vs Profit")

plt.subplot(2, 2, 4)
plt.pie(sales, labels=days, autopct="%1.1f%%", startangle=90)
plt.title("Sales Distribution")

plt.tight_layout()
plt.show()
```
