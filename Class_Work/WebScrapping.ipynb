```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://timely-sunshine-e821b3.netlify.app/'
```

```python
page = requests.get(URL)
page.status_code
```

```python
htmlCode = page.text
print(htmlCode)
```

```python
soup = BeautifulSoup(htmlCode, 'html.parser')
print(soup)
```

```python
content = soup.find_all('div', attrs={'class': 'a'})
for i in content:
    print("Div:{a}", i, end='\n\n\n\n')
```

```python
content = soup.find_all('div', attrs={'class': 'name'})

for item in content:
    text = item.text.strip()
    print(text)
```

```python
content = soup.find('div', attrs={'class': 'price'})
text = content.text.strip()
print(text)
```

```python
content = soup.find_all('div', attrs = {'class': 'image'})
print(content)
```

```python
content = soup.find_all('div', attrs={'class': 'a'})
names = []
prices = []
images = []
for item in content:
    name = item.find('div', attrs={'class': 'name'}).text.strip()
    price = item.find('div', attrs={'class': 'price'}).text.strip()
    image = item.find('div', attrs={'class': 'image'}).img['src']

    names.append(name)
    prices.append(price)
    images.append(image)

print("Names: ", names)
print("Prices: ", prices)
print("Images: ", images)
```

```python
df = pd.DataFrame({'Product_Name': names, 'MRP': prices, 'Image_SRC': images})
df
```

```python
df.to_csv('products.csv', header = True, index=False)
```
